```python
# Problem - 4: Matrix Operations:
A = np.array([[3,4],[7,8]])
B = np.array([[5,3],[2,1]])
# Prove following with Numpy:
# 1. Prove A.A−1 = I.
print("A.A_inv =\n",np.dot(A,np.linalg.inv(A)))
# A.A_inv=identity matrix proved.
```

```python
# 2. Prove AB ̸= BA.
print("AB =\n",np.dot(A,B))
print("BA =\n",np.dot(B,A))
# AB !=BA proved.
```

```python
# 3. Prove (AB)T = BTAT
print("(AB)^T =\n",np.dot(A,B).T)
print("B^T.A^T =\n",np.dot(B.T,A.T))
# (AB)T = BTAT proved.
```

```python
# Solve the following system of Linear equation using Inverse Methods.

# 2x − 3y + z = −1
# x − y + 2z = −3
# 3x + y − z = 9

A2=np.array([[2,-3,1],
             [1,-1,2],
             [3,1,-1]])
B2=np.array([-1,-3,9])

X2=np.linalg.solve(A2,B2)
print("\nUsing linalg.solve function: ",X2)

#Solve using np.linalg.inv function
X=np.dot(np.linalg.inv(A2),B2)
print("\nUsing linalg.inv function: ",X)
```
